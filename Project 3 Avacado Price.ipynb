```python
#applying linear regression 
x = data['Total Volume'].values
y = data['AveragePrice'].values
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

n = len(x)
numer = 0
denom = 0

for i in range(n):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) **2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
print(b1,b0)    
```

```python
max_x = np.max(x) + 100
min_x = np.min(x) - 100

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, color='orange', label='Regression Line')
plt.scatter(x, y, c='blue', label='Scatter Plot')

plt.xlabel('Total Volume')
plt.ylabel('AveragePrice')
plt.legend()
plt.show()
```
